# Arnold's Cat Map

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import imageio
```

```python
# arnold's cat map takes a point (x, y) then 
# chooses a new position for it within the 
# dimensions (xdim, ydim) of the image
def catmap(x, y, xdim, ydim):
    return (x + y) % xdim, (x + 2*y) % ydim
```

```python
# apply catmap to an image one time
def step(img):
    xdim = img.shape[0]
    ydim = img.shape[1]
    new_img = np.array(img)
    
    for x, row in enumerate(img):
        for y, _ in enumerate(row):
            newx, newy = catmap(x, y, xdim, ydim)
            new_img[newx][newy] = img[x][y]
    
    return new_img
```

```python
cat = mpimg.imread('cat.jpg')
plt.imshow(cat)
```

We'll be applying the mapping to the image above

```python
# here we're making a gif of the stages of the mapping

# we want the first few frames to be still so it's easier to see the original image
still_frames = 5
frames = [cat for _ in range(still_frames)]
counter = 1
new_cat = step(cat)
while not (new_cat == cat).all():
    frames.append(new_cat)
    new_cat = step(new_cat)
    counter += 1
imageio.mimsave('cat.gif', frames, 'GIF', duration=0.1)
```
